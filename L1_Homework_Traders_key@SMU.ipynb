### Exercise 1: Data Collection and Basic Operations
### Topics Used: For Loops, Indexing, Lists, Operations, Variables

#### 1. You are given a list of daily closing prices for a stock over 5 days: [100.5, 102.3, 99.8, 103.7, 101.2].

#### 2. Store this list in a variable called stock_prices.
#### 3. Calculate the daily returns for each day (return = (current_price - previous_price) / previous_price) and store them in a new list called daily_returns.
#### 4. Print the daily_returns list.

```python
# Submit Code Here 

# Exercise 1: Calculate daily returns from stock prices

# 1. Store the list of stock prices
stock_prices = [100.5, 102.3, 99.8, 103.7, 101.2]

# 2. Calculate daily returns by tracking previous price
daily_returns = []
previous_price = stock_prices[0]  # Start with first price
for current_price in stock_prices[1:]:  # Iterate from second price onward
    daily_return = (current_price - previous_price) / previous_price
    daily_returns.append(daily_return)  # No rounding
    previous_price = current_price  # Update previous price

# 3. Print the daily returns
print(daily_returns)
```

### Exercise 3: Loops and Trading Signals

#### Using stock_prices, create signals ("Buy" if >2% increase, "Sell" if <-2% decrease, "Hold" otherwise).
#### Break on "Sell".
#### Print signals and break message.

```python
# Submit Code Here 

stock_prices = [100.5, 102.3, 99.8, 103.7, 101.2]
signals = []
previous_price = stock_prices[0]
for current_price in stock_prices[1:]:
    percent_change = (current_price - previous_price) / previous_price * 100
    if percent_change > 2:
        signals.append("Buy")
    elif percent_change < -2:
        signals.append("Sell")
    else:
        signals.append("Hold")
    if signals[-1] == "Sell":
        print("Signals: " + str(signals))
        print("Loop stopped due to Sell signal")
        break
    previous_price = current_price
if signals[-1] != "Sell":
    print("Signals: " + str(signals))
```

### Final Challenge Exercise 5: Object-Oriented Programming - Advanced Trading Strategy Class

#### Build an advanced TradingStrategy class to simulate a stock trading system with detailed tracking and reporting. Follow these steps:

#### 1. Set up the class with:
#### &nbsp;A name for the strategy.
#### &nbsp;A list for stock prices.
#### &nbsp;A list for trading signals.
#### &nbsp;A list to record trades.
#### &nbsp;Variables for cash (start with $10,000) and shares (start with 0).
#### &nbsp;A fixed number of shares to trade each time (e.g., 10).
#### 2. Create a method to add a new stock price to the class.
#### 3. Create a method to generate trading signals:
#### &nbsp;Compare each price to the previous one.
#### &nbsp;Use a 2% increase for "Buy," a 2% decrease for "Sell," and anything else for "Hold."
#### 4. Create a method to execute trades based on signals:
#### &nbsp;For "Buy," purchase shares if cash allows, updating cash and shares.
#### &nbsp;For "Sell," sell all shares if any are owned, updating cash and shares.
#### &nbsp;Record each trade as a tuple with price, action, and shares traded.
#### 5. Create a method to calculate the total portfolio value using the latest price.
#### 6. Create a method to provide a summary including:
#### &nbsp;Strategy name, signals, trades, final cash, shares owned, and total portfolio value.
#### 7. Test your class:
#### &nbsp;Use the prices [100.5, 102.3, 99.8, 103.7, 101.2].
#### &nbsp;Add each price, generate signals, execute trades, and print the summary.


```python
# Submit Code Here 

# Exercise 5: Enhanced TradingStrategy class

class TradingStrategy:
    def __init__(self, name):
        self.name = name
        self.prices = []          # List to store stock prices
        self.signals = []         # List to store trading signals
        self.trades = []          # List to store trade records as tuples
        self.cash = 10000         # Initial cash balance
        self.shares = 0           # Initial shares owned
        self.lot_size = 10        # Number of shares per trade
    
    def add_price(self, price):
        self.prices.append(price)  # Add a new price to the list
    
    def generate_signals(self):
        previous_price = self.prices[0]  # Start with first price
        # Generate signals starting from second price
        for current_price in self.prices[1:]:
            percent_change = (current_price - previous_price) / previous_price * 100
            if percent_change > 2:
                self.signals.append("Buy")
            elif percent_change < -2:
                self.signals.append("Sell")
            else:
                self.signals.append("Hold")
            previous_price = current_price  # Update previous price
    
    def execute_trades(self):
        current_prices = self.prices[1:]  # Prices starting from day 2
        for signal in self.signals:
            current_price = current_prices[0]  # Get next price
            if signal == "Buy" and self.cash >= current_price * self.lot_size:
                self.shares = self.shares + self.lot_size
                self.cash = self.cash - current_price * self.lot_size
                trade = (current_price, "Buy", self.lot_size)
                self.trades.append(trade)
            elif signal == "Sell" and self.shares > 0:
                sell_shares = self.shares  # Sell all owned shares
                self.cash = self.cash + current_price * self.shares
                self.shares = 0
                trade = (current_price, "Sell", sell_shares)
                self.trades.append(trade)
            elif signal == "Sell" and self.shares == 0:
                # Log a "Sell" with 0 shares if no position to close
                trade = (current_price, "Sell", 0)
                self.trades.append(trade)
            current_prices = current_prices[1:]  # Move to next price
    
    def portfolio_value(self):
        latest_price = self.prices[-1]  # Use last price
        total_value = self.cash + self.shares * latest_price
        return total_value
    
    def summary(self):
        output = self.name + " Summary:\n"
        output = output + "Signals: " + str(self.signals) + "\n"
        output = output + "Trades: " + str(self.trades) + "\n"
        output = output + "Final Cash: $" + str(self.cash) + "\n"
        output = output + "Shares Owned: " + str(self.shares) + "\n"
        output = output + "Total Portfolio Value: $" + str(self.portfolio_value())
        return output

# Test the class
strategy = TradingStrategy("Momentum Strategy")
prices = [100.5, 102.3, 99.8, 103.7, 101.2]
for price in prices:
    strategy.add_price(price)

strategy.generate_signals()
strategy.execute_trades()
print(strategy.summary())
```
